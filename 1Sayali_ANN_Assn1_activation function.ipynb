# 1. Sigmoid Function and Derivative of Sigmoid Function

```python
def sigmoid(x):
    return (1/(1+ np.exp(-x)))
```

```python
x = np.linspace(10,-10,100)
y = sigmoid(x)
```

```python
plt.figure()
plt.plot(x,y)
plt.title("Activation Function : Sigmoid")
plt.show()
```

```python
def der_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
x = np.linspace(10,-10,100)
y = sigmoid(x)
dy = der_sigmoid(x)

plt.plot(x,y,x,dy)
plt.axis('tight')
plt.legend(['sigmoid' , 'der_sigmoid'])
plt.title("Activation Function : Derivative of Sigmoid")
plt.show()
```

# 2. Tanh and TanH Shrink

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.linspace(10,-10,100)
y = np.tanh(x)

plt.figure()
plt.plot(x,y)
plt.show()
```

```python
A = x-np.tanh(x)

plt.figure()
plt.plot(x,A)
plt.show()
```

# 12. ElU

```python
a = 1.0
def elu(x):
    if x > 0:
        return x
    else:
        return a*(np.exp(x) - 1)
result = [] 
for i in x:
    result.append(elu(i))
plt.plot(x,result)    
```

# 15. GeLU
